## Working with Missing Data in Pandas

```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

### Filling missing values using fillna(), replace() and interpolate()

```python
data = {'names': ['steve', 'john', 'richard', 'sarah', 'randy', 'micheal', 'julie'],
        'age': [20,22,20,21,24,23,22],
        'gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'rank': [2,1,4,5,3,7,6]}
```

```python
ranking_df = DataFrame(data)
ranking_df.iloc[2:5, 1] = np.nan
ranking_df.iloc[3:6, 3] = np.nan
ranking_df.iloc[3,:] = np.nan
ranking_df
#setting specific groups of columns and rows to NaN
```

```python
ranking_df.isnull()
#printing df with all values null set to true and false otherwise
```

```python
ranking_df.notnull()
#printing df with all values not null set to true and false otherwise
```

```python
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
#only returning rows where age is missing
```

```python
#filling a missing value with a single value
ranking_df.fillna(0)
```

```python
#fill missing values with value that comes prior to missing value
ranking_df.fillna(method='pad')
```

```python
#fill missing value with next value that is not missing
ranking_df.fillna(method='bfill')
```

```python
#fill missing value using linear method which ignores the index and treats the vals as equally spaced
ranking_df.interpolate(method='linear')
```

```python
#drop all rows and cols that have missing values
ranking_df.dropna()
```

```python
#drop all rows and cols that only contains missing values
ranking_df.dropna(how='all')
```

```python
#drop all cols which contain at least one missing value
ranking_df.dropna(axis=1)
```

```python
#drop all rows that have at least one missing value
ranking_df.dropna(axis=0)
```
